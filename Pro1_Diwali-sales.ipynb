Gender

```python
#Plot a bar graph for gender
sns.countplot(data=df, x="Gender")  #>>2 parameters
#Axes level
#Figure level
```

From the Above graph we can see that females are purchasing more compare to Males

```python
df["Gender"].value_counts()
```

```python
#Who has spent more (Amount >> Female >> male)
sales_gen = df.groupby("Gender")["Amount"].sum().reset_index()
sales_gen
#Result of groupby is always a series not a dataframe b4 reset_index
```

```python
sns.barplot(data=sales_gen,x="Gender",y="Amount")  #>> 3 parameters
```


From the above graph we can conclude that Females have spent more money than Males

Age group

```python
#Age group frequency
df["Age Group"].value_counts().head(3)
```

```python
sns.countplot(data=df,x="Age Group",hue='Gender',palette='viridis') #hue >> to bifurcate more deep dive like here we differentiated the age group by gender wise
```

By the above graph we can see the age group between 26-35 are having more frequency , gradually increasing till the age 45 and then later the frequency is decreasing

```python
#Amount spent by age group
sales_age_grp = df.groupby("Age Group")["Amount"].sum().reset_index()
sales_age_grp
```

```python
sns.barplot(data=sales_age_grp,x="Age Group",y="Amount",palette='viridis')
```



From the above graph we can conclude that the age group 26 - 35 arre spending more money compare to any other group

State

```python
df["State"].value_counts().head(10)
```

```python
#Orders placed with respect to the states
sales_states_ord = df.groupby('State')['Orders'].sum().sort_values(ascending=False).head(10).reset_index()
sales_states_ord
```

```python
sns.barplot(data=sales_states_ord, x= "State",y = "Orders",palette = "viridis")
plt.xticks(rotation = 90)
```

As we can see from the above graph that uttar Pradesh is the highest order placing state then Maharashtra and Karnataka as the top 3 values for the same

```python
#Amount placed with respect to the states
sales_state_amt = df.groupby("State")["Amount"].sum().sort_values(ascending= False).head(10).reset_index()
sales_state_amt
```

```python
sns.barplot(data=sales_state_amt, x= "State", y = "Amount" ,palette = "deep")
plt.xticks(rotation=90)
```

As we have observed earlier that uttar Pradesh is the highest order placing state so correspondingly its also the highest state who is spending more money compare to other states.
